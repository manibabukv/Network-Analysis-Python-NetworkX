```python
df = pd.read_excel(r'/Users/kvmanibabu/Desktop/RA_Project/Network_Analysis_Final_Dataset.xlsx')
df.head()
```

```python
G = nx.from_pandas_edgelist(df, source='Cause', target='Effect', create_using=nx.DiGraph)
nx.draw(G)
```

```python
net = Network(notebook=True,height='100%', width='100%' ,directed = True)
net.from_nx(G)
net.show('networkanalysisssfinal.html')
```

```python
net = Network(height='400px', width='50%',directed=True,font_color='#ff8000',bgcolor='606060')
net.from_nx(G)
net.show("example.html")
```
